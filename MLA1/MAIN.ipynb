```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("training_dataset.csv")
df.head(-1)
```

```python
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
df.duplicated()[df.duplicated()==True]
```

```python
print("Shape (rows, columns): ", df.shape,"\n")
print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
df.dtypes
```

```python
df['age'] = df['age'].replace(r'', np.nan, regex=True)

```

```python
print(df)
```

```python
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['income'], bins=20)
plt.show()
```

```python
city_revenue = df.groupby('city')['income'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
city_revenue.plot(kind='bar')
plt.show()

city_revenue
```

```python
city_revenue = df.groupby('age')['income'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
city_revenue.plot(kind='bar')
plt.show()

city_revenue
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='score', data=df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['age', 'income']].corr(), annot=True)
plt.show()
```

```python
monthly_revenue = df.groupby('age')['score'].sum()

plt.figure(figsize=(10,5))
monthly_revenue.plot()
plt.show()
```
